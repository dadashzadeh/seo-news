```python
# Importing necessary libraries
from GoogleNews import GoogleNews

# Setting up the parameters for GoogleNews search
googlenews = GoogleNews()
googlenews.set_period('1d')
googlenews.set_lang('en')
googlenews.set_encode('utf-8')

# keyword GoogleNews search
googlenews.search('SEO')

# Creating a dataframe from the results of the GoogleNews search
data = pd.DataFrame.from_dict(googlenews.results())
data = data.drop(columns=["img", "date", "media"])
data['datetime'] = pd.to_datetime(data['datetime']).dt.strftime('%Y/%m/%d %H:%M:%S')
```

```python
import feedparser
```

```python
# List of RSS feeds related to SEO
lists_rss_seo = ["http://feeds.seroundtable.com/SearchEngineRoundtable1",
                 "https://www.searchenginejournal.com/category/seo/feed/",
                 "https://searchengineland.com/feed",
                 "https://moz.com/posts/rss/blog", ]

# Function to convert the time format of the RSS feed entries
def time_rss(input_time: str):
    try:
        dt = datetime.datetime.strptime(input_time, '%a, %d %b %Y %H:%M:%S %z')
    except:
        dt = datetime.datetime.fromisoformat(input_time)
    return dt.strftime('%Y/%m/%d %H:%M:%S')

# Function to strip HTML tags from the RSS feed entries
def strip_html(s):
    return html.fromstring(s).text_content().strip()

# Creating a dataframe from the RSS feed entries
df = pd.DataFrame(columns=["title", "datetime", "desc", "link"])
for list_rss in lists_rss_seo:
    d = feedparser.parse(list_rss)
    for datarss in d['entries']:
        link = datarss["link"]
        title = datarss["title"].strip()
        summary = strip_html(datarss["summary"])
        published = time_rss(datarss["published"])
        df.loc[len(df)] = [title, published, summary, link]
```
